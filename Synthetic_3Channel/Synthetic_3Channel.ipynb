```python
from pathlib import Path
import sys
sys.path.append('../')
from pma_open import *

#My image
file_path = '../Dropbox Files/hel1.pma'
image_path = "../Channel Mapping/hel1_Avg_Frame/hel1_Avg_Frame.png"
image = io.imread(image_path, as_gray=True)
```

```python
three_CH_hel1 = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
fig, ax = plt.subplots()
plt.imshow(three_CH_hel1, cmap="grey")
plt.grid(())
print("This should just be an all black image of shape (512, 512)")
print(f"Image Shape: {three_CH_hel1.shape}")
```

# Simple CH3

```python
# Duplicating CH2 to populate CH3 
three_CH_hel1[:, :171] = image[:,:171]
three_CH_hel1[:,171:342] = image[:,256:427]
three_CH_hel1[:,342:] = image[:,256:426]

fig, ax = plt.subplots()
plt.imshow(three_CH_hel1, cmap="grey")
plt.axvline(x= 171, color='w', linestyle='--', linewidth=1)
plt.axvline(x= 342, color='w', linestyle='--', linewidth=1)
plt.title("Simple Synthetic 3 Channel Image")

image_path = "Synthetic Images hel1/Simple_3CH_hel1.png"
CH1_image_path = "Synthetic Images hel1/Simple_3CH_hel1_CH1.png"
CH2_image_path = "Synthetic Images hel1/Simple_3CH_hel1_CH2.png"
CH3_image_path = "Synthetic Images hel1/Simple_3CH_hel1_CH3.png"

avg_frame_data_CH1 = three_CH_hel1[:,:171]
avg_frame_data_CH2 = three_CH_hel1[:,171:342]
avg_frame_data_CH3 = three_CH_hel1[:,342:]

# Image.fromarray(three_CH_hel1).save("Synthetic Images hel1/Simple_3CH_hel1.png")
# Image.fromarray(avg_frame_data_CH1).save(CH1_image_path)
# Image.fromarray(avg_frame_data_CH2).save(CH2_image_path)
# Image.fromarray(avg_frame_data_CH3).save(CH3_image_path)
```

```python
#Chanel arrays
print(f"Average CH1 Frame Data- shape: {avg_frame_data_CH1.shape}")
print(f"Average CH1 Frame Data- Type: {type(avg_frame_data_CH1)}")
print(f"Average CH1 Frame Data- dType: {avg_frame_data_CH1.dtype}")

print(f"\nAverage CH2 Frame Data- shape: {avg_frame_data_CH2.shape}")
print(f"Average CH2 Frame Data- Type: {type(avg_frame_data_CH2)}")
print(f"Average CH2 Frame Data- dType: {avg_frame_data_CH2.dtype}")

print(f"\nAverage CH3 Frame Data- shape: {avg_frame_data_CH3.shape}")
print(f"Average CH3 Frame Data- Type: {type(avg_frame_data_CH3)}")
print(f"Average CH3 Frame Data- dType: {avg_frame_data_CH3.dtype}")
```
